### 1개의 독립변수와 종속변수 mpg 와의 관계

#### (cylinders, mpg)

````python
sns.regplot(data=mpg, x='cylinders', y='mpg')
````

#### (displacement, mpg)

````python
sns.regplot(data=mpg, x='displacement', y='mpg')
````

#### (horsepower, mpg)

````python
sns.regplot(data=mpg, x='horsepower', y='mpg')
````

#### (weight, mpg)

````python
sns.regplot(data=mpg, x='weight', y='mpg')
````

#### (acceleration, mpg)

````python
sns.regplot(data=mpg, x='acceleration', y='mpg')
````

#### (model_year, mpg)

````python
sns.regplot(data=mpg, x='model_year', y='mpg')
````

```
위 그래프들을 보고, 각 변수에 대해 서술해보면 다음과 같다.
cylinders(배기통): 회귀선의 기울기가 음수이므로, cylinders 가 증가할수록 mpg가 감소하는 경향을 보임.
displacement (배기량): 회귀선의 기울기가 음수이므로, displacement이 증가할수록 mpg가 감소하는 경향을 보임.
horsepower (출력): 회귀선의 기울기가 음수이므로, horsepower가 증가할수록 mpg가 감소하는 경향을 보임.
weight (무게): 회귀선의 기울기가 음수이므로, weight가 증가할수록 mpg가 감소하는 경향을 보임.
acceleration (가속도): 가속도는 상대적으로 mpg에 미치는 영향이 적어 보임.
model_year (모델 연도): 회귀선의 기울기가 양수이므로, model_year가 증가할수록 mpg도 증가하는 경향을 보임.
따라서, 주어진 그래프를 분석한 결과, displacement, horsepower, weight, model_year 변수 중 weight (무게) 변수가 mpg에 가장 큰 
영향을 미치는 것으로 나타납니다. weight 변수의 회귀선이 기울기가 음수이고, 산점도에서 weight 값이 증가할수록 mpg 값이 감소하는 
경향을 보이기 때문임.
```

### 여러개의 독립변수와 종속변수 mpg 의 관계

````python
sns.pairplot(data=mpg)
````

````python
print(mpg.corr(method='pearson'))
````

```
위의 그래프들 보고 종합적으로 분석해보면, cylinders, displacement, horsepower, weight 는 연비와 모두 음의 상관관계를 갖고 acceleration, model_year은  양의 상관관계를 갖음을 알 수 있다.
또한 상관계수를 분석한 자료를 보면, 연비와 weight 사이의 피어슨 상관계수가 제일 크므로 연비에 가장 영향을 크게 미치는 변수는 
weight 임을 알 수 있다.
mpg 예측을 위한 최종모형을 구축하기 위한 전략은 반복적인 분석을 통해 영향을 주는 가장 큰 변수들을 하나씩 줄여나가 최종적으로 
영향을 가장 적게 주는 변수들만 남겨놓는 것이다.
```
