## 2. 데이터 연결하기

### concat() 함수를 이용한 데이터프레임 연결하기

```python
df1 = pd.DataFrame({'A': ['a0', 'a1', 'a2', 'a3'],
                    'B': ['b0', 'b1', 'b2', 'b3'],
                    'C': ['c0', 'c1', 'c2', 'c3'],
                    'D': ['d0', 'd1', 'd2', 'd3']})
df1
```

```python
df2 = pd.DataFrame({'A': ['a4', 'a5', 'a6', 'a7'],
                    'B': ['b4', 'b5', 'b6', 'b7'],
                    'C': ['c4', 'c5', 'c6', 'c7'],
                    'D': ['d4', 'd5', 'd6', 'd7']})
df2
```

```python
df3 = pd.DataFrame({'A': ['a10', 'a11', 'a12', 'a13'],
                    'B': ['b10', 'b11', 'b12', 'b13'],
                    'C': ['c10', 'c11', 'c12', 'c13'],
                    'D': ['d10', 'd11', 'd12', 'd13']})
df3
```

#### 행방향으로 데이터프레임 연결하기 axis=0 (default)

```python
row_concat = pd.concat([df1, df2, df3])
row_concat
```

```python
row_concat.iloc[3]
```

```python
row_concat.loc[3]
```

#### 데이터프레임과 시리즈를 연결하면 이름이 없는 열이 되므로 뜻하지 않은 결과를 얻는다

```python
new_row_series = pd.Series(['n1', 'n2', 'n3', 'n4'])
new_row_series
```

```python
new_row_series.shape
```

```python
new_row_series.ndim
```

```python
pd.concat([df1, new_row_series])
```

```python
new_row_df = pd.DataFrame([['n1', 'n2', 'n3', 'n4']], columns = ['A', 'B', 'C', 'D'])
new_row_df
```

```python
new_row_df.shape
```

```python
pd.concat([df1, new_row_df])
```

```python
df1.append(new_row_df)
```

```python
pd.concat([df1, new_row_df], ignore_index = True)
```

#### 열방향으로 데이터프레임 연결하기 axis = 1

```python
col_concat = pd.concat([df1, df2, df3], axis = 1)
col_concat
```

```python
col_concat.drop(0)
```

```python
col_concat.drop('D', axis = 1)
```

```python
pd.concat([df1, df2], axis = 1, ignore_index = True)
```

#### 행과 열 index가 일부가 일치하지 않은 데이터프레임 연결하기

```python
df2.columns = ['A', 'B', 'E', 'F']
df2
```

```python
row_concat = pd.concat([df1, df2])
row_concat
```

```python
pd.concat([df1, df2], join = 'inner', ignore_index = True)
```

### Merge()  함수를 이용한 데이터프레임 연결

```python
df1 = pd.DataFrame({'key':['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'data1':range(7)})
df1
```

```python
df2 = pd.DataFrame({'key':['a', 'b', 'd'], 'data2':range(3)})
df2
```

```python
pd.merge(df1, df2, on = 'key')
```

```python
df1.merge(df2, on = 'key')
```

```python
pd.merge(df1, df2, how = 'outer')
```

```python
pd.merge(df1, df2, how = 'left')
```
